## 문제1
- 결측치 데이터(행)을 제거하고, 앞에서부터 70% 데이터만 활용해, 'f1' 컬럼 1사분위 값을 구하시오
- 데이터: members.csv

```python
import pandas as pd
df = pd.read_csv("members.csv")
#print(df)
df.isnull().sum()
df = df.dropna()
#print(df)

#print(int(len(df)*0.7))
df= df.iloc[:int(len(df)*0.7)]
print(df.shape)

print(df['f1'].quantile(.25))
```

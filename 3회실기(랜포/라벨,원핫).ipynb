## 데이터 전처리 및 피처엔지니어링

```python
# 수치형 데이터와 범주형 데이터 분리 
c_train= train.select_dtypes(include='O').copy()
c_test = test.select_dtypes(include='O').copy()
n_train = train.select_dtypes(exclude='O').copy()
n_test = test.select_dtypes(exclude='O').copy()
```

```python
# 수치형 변수 스케일링
c_train
```

```python
# 범주형 변수 인코딩 (더미)
c_train = pd.get_dummies(c_train)
c_test = pd.get_dummies(c_test)
c_train.head()

```

```python
cols = train.select_dtypes(include='O').columns
from sklearn.preprocessing import LabelEncoder
for col in cols:
  le = LabelEncoder()
  train[col] = le.fit_transform(train[col])
  test[col] = le.transform(test[col])

train[cols]
```

```python
train
```

```python
# 분리한 데이터 다시 합침
train = pd.concat([n_train,c_train],axis=1)
test = pd.concat([n_test,c_test],axis=1)

train.head()
```

```python
train= train.drop('Unnamed: 0',axis=1)
test = test.drop('Unnamed: 0',axis=1)
```

## 검증데이터 분리

```python
# 검증데이터 분리
from sklearn.model_selection import train_test_split
X_tr,X_val,y_tr,y_val = train_test_split(train.drop('TravelInsurance',axis=1),train['TravelInsurance'],test_size=0.1,random_state=1204)
X_tr.shape,X_val.shape,y_tr.shape,y_val.shape
```

## 모델 학습 및 예측

```python
# 랜덤포레스트
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=400,max_depth=5,random_state=1204)
rf.fit(X_tr, y_tr)
pred = rf.predict_proba(X_val)[:,1]

```

```python
# 평가
from sklearn.metrics import roc_auc_score
roc_auc_score(y_val, pred)
```
